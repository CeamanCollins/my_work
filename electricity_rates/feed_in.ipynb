# Feed In

This notebook calculates the amount earned through feeding electricity back into the network at different rates for two properties: olr and 1mc. 

```python
# Importing packages
import pandas as pd
import matplotlib.dates as mdates
```

```python
# Importing data
data_1mc = pd.read_csv('data/HDF_calckWh_10000826638_26-02-2025.csv', parse_dates=['Read Date and End Time'], dayfirst=True, index_col='Read Date and End Time', usecols=['Read Type', 'Read Value', 'Read Date and End Time'])
data_olr = pd.read_csv('data/HDF_calckWh_10000847339_26-02-2025.csv', parse_dates=['Read Date and End Time'], dayfirst=True, index_col='Read Date and End Time', usecols=['Read Type', 'Read Value', 'Read Date and End Time'])
data_1mc.index.rename('Date', inplace=True)
data_olr.index.rename('Date', inplace=True)
```

```python
export_data_1mc = data_1mc[data_1mc['Read Type'] == 'Active Export Interval (kWh)']
export_data_olr = data_olr[data_olr['Read Type'] == 'Active Export Interval (kWh)']

tariff_1_1mc = export_data_1mc[(export_data_1mc.index<='2025-01-12')]['Read Value'].sum()
tariff_2_1mc = export_data_1mc[(export_data_1mc.index>'2025-01-12')]['Read Value'].sum()

tariff_1_olr = export_data_olr[(export_data_olr.index<='2025-01-12')]['Read Value'].sum()
tariff_2_olr = export_data_olr[(export_data_olr.index>'2025-01-12')]['Read Value'].sum()   
 
rate_1 = 0.24  # €/kWh rate for period 1
rate_2 = 0.20  # €/kWh rate for period 2

total_olr = tariff_1_olr * rate_1 + tariff_2_olr * rate_2

total_1mc = tariff_1_1mc * rate_1 + tariff_2_1mc * rate_2

print(f'Total for olr: €{total_olr:.2f}')
print(f'Total for 1mc: €{total_1mc:.2f}')
```
